```python
def small_calculator_function(number1,number2): # defining
    product = number1 * number2
    difference = number1 - number2
    addition = number1 + number2
    return [addition,difference,product]
```

```python
small_calculator_function(10,5) # calling
```

```python
small_calculator_function(100,500)
```
